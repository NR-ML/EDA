```python
from sklearn import datasets
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind ="scatter", x ='sepal.length', y ='petal.length')
plt.grid()
```

```python
ax = df[df.variety=="Setosa"].plot.scatter(x='sepal.length', y='petal.length', color='red', label='setosa')
df[df.variety=="Versicolor"].plot.scatter(x='sepal.length', y='petal.length', color='green', label='versicolor', ax=ax)
df[df.variety=="Virginica"].plot.scatter(x='sepal.length', y='petal.length', color='blue', label='virginica', ax=ax)
ax.set_title("scatter")
```

```python
ax = df[df.variety=="Setosa"].plot.scatter(x='sepal.width', y='petal.width', color='red', label='setosa')
df[df.variety=="Versicolor"].plot.scatter(x='sepal.width', y='petal.width', color='green', label='versicolor', ax=ax)
df[df.variety=="Virginica"].plot.scatter(x='sepal.width', y='petal.width', color='blue', label='virginica', ax=ax)
ax.set_title("scatter")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
  
def graph(y):
    sns.boxplot(x="variety", y=y, data=df)
  
plt.figure(figsize=(10,10))
      
# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph('sepal.length')
  
plt.subplot(222)
graph('sepal.width')
  
plt.subplot(223)
graph('petal.length')
  
plt.subplot(224)
graph('sepal.width')
  
plt.show()
```

```python
plot = sns.FacetGrid(df, hue="variety")
plot.map(sns.distplot, "sepal.length").add_legend()
  
plot = sns.FacetGrid(df, hue="variety")
plot.map(sns.distplot, "sepal.width").add_legend()
  
plot = sns.FacetGrid(df, hue="variety")
plot.map(sns.distplot, "petal.length").add_legend()
  
plot = sns.FacetGrid(df, hue="variety")
plot.map(sns.distplot, "petal.width").add_legend()
  
plt.show()
```
